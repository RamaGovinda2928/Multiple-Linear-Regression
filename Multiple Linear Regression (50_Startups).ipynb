```python
x = df.drop(['Profit'], axis = 1).values
y = df.iloc[:, 3].values.reshape(-1,1)
```

```python
# Model Fitting

import statsmodels.formula.api as smf
from sklearn import linear_model
from sklearn.metrics import r2_score

```

```python
model1 = linear_model.LinearRegression()
model1.fit(x,y)
```

```python
# Coefficients and intercept

model1.coef_
```

```python
model1.intercept_
```

```python
# R2 Score

r2_score(y,model1.predict(x))
```

```python
df_new=pd.DataFrame({"R&D Spend":152200,"Administration":155300,"Marketing Spend":472000},index=[1])
model1.predict(df_new)
print("The profit is:",model1.predict(df_new))
```

```python
d1=df.rename({'Marketing Spend':'Marketing_Spend'},axis=1)
df1=d1.rename({'R&D Spend':'RandD_Spend'},axis=1)

```

```python
df1.head(3)
```

```python
module_ = smf.ols("Profit~RandD_Spend+Administration+Marketing_Spend",data = df1).fit()
module_.summary()
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(module_.resid,line='q')
```

```python
#Residual Plot for Homoscedasticity

def std( vals ):
    return (vals - vals.mean())/vals.std()

plt.scatter(std(module_.fittedvalues),
            std(module_.resid))
```

```python
fig = plt.figure(figsize=(17,10))
fig = sm.graphics.plot_regress_exog(module_, "RandD_Spend",fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(17,10))
fig = sm.graphics.plot_regress_exog(module_, "Administration", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(17,10))
fig = sm.graphics.plot_regress_exog(module_, "Marketing_Spend", fig=fig)
plt.show()
```

```python
#Model Deletion Diagnostics
module_influence = module_.get_influence()
(c, _) = module_influence.cooks_distance
```

```python
# influence plot

from statsmodels.graphics.regressionplots import influence_plot
influence_plot(module_)
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```

```python
#Since the value is <1 , we doesn't need to do the diagnostic process and finalize the model
(module_.rsquared,module_.aic)
```
